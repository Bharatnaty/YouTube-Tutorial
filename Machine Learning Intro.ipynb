# 1. MEASURE OF POSITION

### PERCENTILE

```python
import numpy as np
import pandas as pd
num=list(np.random.randint(100,1000,(10)))
print("Randomly Generated Number",num)
sort_num=sorted(num)
print("Sorting",sort_num)
```

```python
# find percentile for number in 3rd position
# (no of value below X/ total n)* 100 => percentile
(2/10)*100
```

```python
# To find number in 50th percentile 
# (50/100)*n
pos=int((50/100)*10)
print(pos)
print("value of 50th percentile",sort_num[pos])
print("it means 50 percentage of value less than the value",sort_num[pos])
```

### QUANTILE

```python
# quantiles (25,50,75,100 percentiles) 
print("25 quantiles",np.percentile(num,25))
print("50 quantiles",np.percentile(num,50))
print("75 quantiles",np.percentile(num,75))
```

### Z-SCORE

```python
# to calculate the Z score = (x-mean)/std
```

```python
from scipy import stats
print(num)
print(stats.zscore(num))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
num=[22,33,33,44]
sns.distplot(num)
plt.title("GAUSSIAN NORMAL DISTRIBUTION")
plt.show()
```

```python
sns.distplot(stats.zscore(num))
plt.title("STANDARD NORMAL DISTRIBUTION")
plt.show()
```

### MEDIAN

```python
median_odd=list(np.random.randint(10,60,9))
median_even=list(np.random.randint(10,60,10))
```

```python
median_odd
```

```python
# if you have n numbers as even ,then formula= ((n/2)+(n+1/2))/2
# if you have n numbers as odd ,then formula= (n+1/2)

# FOR EVEN NUMBERS OF DATA
median_odd=sorted(median_odd)
print(sorted(median_odd))
print(int((len(median_odd)+1)/2))
print("MEDIAN ODD ",median_odd[int((len(median_odd)+1)/2)])

# FOR ODD NUMBER OF DATA
median_even=sorted(median_even)
print("POSITION MID 1",int((len(median_even)/2)),"VALUE",median_even[int((len(median_even)/2))])
print("POSITION MID 2",int(((len(median_even)+1)/2)+1),"VALUE",median_even[int(((len(median_even)+1)/2)+1)])

print("MEDIAN EVEN",median_even[int((len(median_even)/2))]+median_even[int(((len(median_even)+1)/2)+1)]/2)
```

# CENTRAL LIMIT THEOREM

The central limit theorem states that if you have a population with mean  and standard deviation  and take sufficiently large random samples from the population with replacement , then the distribution of the sample means will be approximately normally distributed

### step1: generate the population

```python
randint(1, 7, 50)
```

### step2: generate the multiple sample from population

```python
for _ in range(1000):    # it generate the sample 1000 times
    print((randint(1, 7, 50))) 
```

### step3: find mean for each sample's

```python
[mean(randint(1, 7, 50)) for _ in range(1000)]
```
